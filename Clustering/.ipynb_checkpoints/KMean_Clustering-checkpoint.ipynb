# K-Mean Clustering

Dividing the customers of the mall into different groups depending upon thier spending score and find the target group. 

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('Data_set/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

### Applying Elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
cost = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cost.append(kmeans.inertia_)
plt.figure(figsize = (12, 8))
plt.plot(range(1, 11), cost)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('cost')
plt.show()
```

### Fitting the model with optimal number of clusters

```python
kmeansop = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
Y_pred = kmeansop.fit_predict(X)
```

### Visualizing the Clusters

```python
plt.figure(figsize = (12, 8))
plt.scatter(X[Y_pred == 0, 0], X[Y_pred == 0, 1], s = 70, marker = 'x', color = 'red', label = 'Carefull')
plt.scatter(X[Y_pred == 1, 0], X[Y_pred == 1, 1], s = 70, marker = 'x', color = 'green', label = 'Standard')
plt.scatter(X[Y_pred == 2, 0], X[Y_pred == 2, 1], s = 70, marker = 'x', color = 'blue', label = 'Target')
plt.scatter(X[Y_pred == 3, 0], X[Y_pred == 3, 1], s = 70, marker = 'x', color = 'yellow', label = 'Careless')
plt.scatter(X[Y_pred == 4, 0], X[Y_pred == 4, 1], s = 70, marker = 'x', color = 'orange', label = 'Sensible')
plt.scatter(kmeansop.cluster_centers_[:, 0], kmeansop.cluster_centers_[:, 1], s = 100, color = 'black', label = 'Centroids')
plt.title('Cluster of Customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending Score (1 - 100)')
plt.legend()
plt.show()
```
