# Hierarchical Clustering

Dividing the customers of the mall into different groups depending upon thier spending score and find the target group.

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('Data_set/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

### Applying the Dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

### Fitting the model with optimal number of clusters

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
Y_hc = hc.fit_predict(X)
```

### Visualizing the Clusters 

```python
plt.figure(figsize = (12, 8))
plt.scatter(X[Y_hc == 0, 0], X[Y_hc == 0, 1], s = 70, marker = 'x', color = 'red', label = 'Carefull')
plt.scatter(X[Y_hc == 1, 0], X[Y_hc == 1, 1], s = 70, marker = 'x', color = 'green', label = 'Standard')
plt.scatter(X[Y_hc == 2, 0], X[Y_hc == 2, 1], s = 70, marker = 'x', color = 'blue', label = 'Target')
plt.scatter(X[Y_hc == 3, 0], X[Y_hc == 3, 1], s = 70, marker = 'x', color = 'yellow', label = 'Careless')
plt.scatter(X[Y_hc == 4, 0], X[Y_hc == 4, 1], s = 70, marker = 'x', color = 'orange', label = 'Sensible')
plt.title('Cluster of Customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending Score (1 - 100)')
plt.legend()
plt.show()
```
