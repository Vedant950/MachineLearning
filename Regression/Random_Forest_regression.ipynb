# Random Forest Regression

Fitting the best fit random forest regression model on the given position Vs salary dataset and predict the salary of the employees based on the position level in the organization.

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('Data_set/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
Y = dataset.iloc[:, 2].values
```

### Fitting the model

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators = 100, random_state = 0)
reg.fit(X, Y)
```

### Visualizing the dataset

```python
plt.figure(figsize = (12, 8))
plt.scatter(X, Y, color = 'red', marker = 'x')
plt.title('Position Vs Salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

### Visualizing the model

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.figure(figsize = (12, 8))
plt.scatter(X, Y, color = 'red', marker = 'x')
plt.plot(X_grid, reg.predict(X_grid))
plt.title('Position Vs Salary')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
