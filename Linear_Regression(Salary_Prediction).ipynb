# Simple Linear Regression

Fitting the best fit linear regression model in the given Salary_Dataset using sklearn

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset=pd.read_csv('C:/Users/KIIT/Desktop/ML_models/Data_set/Salary_Data.csv')
X=dataset.iloc[:, :-1].values
Y=dataset.iloc[:, 1].values
```

### Splitting the dataset

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, Y_train, Y_test = tts(X, Y, test_size=1/3, random_state=0)
```

### Fitting the Linear regression model

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train, Y_train)
```

### Predicting the test set results

```python
Y_pred=reg.predict(X_test)
```

### Visualizing the dataset

```python
plt.figure(figsize=(10, 8))
plt.scatter(X, Y, color='blue', marker='x')
plt.title('Experience Vs Salary')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

### Visualizing the training set results

```python
plt.figure(figsize=(10,8))
plt.scatter(X_train, Y_train, color='red', marker='x')
plt.plot(X_train, reg.predict(X_train), color='blue')
plt.title('Experience Vs Salary')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

### Visualizing the test set results

```python
plt.figure(figsize=(10,8))
plt.scatter(X_test, Y_test, color='green', marker='x')
plt.plot(X_train, reg.predict(X_train), color='blue')
plt.title('Experience Vs Salary')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
