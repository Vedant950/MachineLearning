# Polynomial Regression

Fitting the best fit polynomial regression model on the data set of Position Vs Salary and predict the salary of the employee depending on their current position in the organization.

### Importing the libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Imporing the dataset 

```python
dataset = pd.read_csv('Data_set/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
Y = dataset.iloc[:, 2].values
```

### Fitting the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
reg = LinearRegression()
reg.fit(X_poly, Y)
```

### Visualizing the dataset

```python
plt.figure(figsize = (12, 8))
plt.scatter(X, Y, color = 'red', marker = 'x')
plt.title('Position Vs Salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

### Visualizing the model 

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.figure(figsize = (12, 8))
plt.scatter(X, Y, color = 'red', marker = 'x')
plt.plot(X_grid, reg.predict(poly_reg.fit_transform(X_grid)))
plt.title('Position Vs Salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
